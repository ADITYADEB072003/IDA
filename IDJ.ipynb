    Age group 20-30: 13 employees
    Age group 30-40: 24 employees
    Age group 40-50: 6 employees
    Age group 50-60: 4 employees

Write a Python script to draw a histogram that displays this distribution of employees by age group.**bold text**

```python
import matplotlib.pyplot as plt

# Age groups (bin edges)
bins = [20, 30, 40, 50, 60]

# Number of employees in each age group
employees = [13, 24, 6, 4]

# Plotting the histogram
plt.hist(, bins=bins, weights=employees, edgecolor='black', rwidth=0.8)
# plt.bar(age_group,employees)
# Adding labels and title
plt.xlabel('Age Groups')
plt.ylabel('Number of Employees')
plt.title('Histogram of Employee Age Groups')
plt.xticks(bins)  # Setting x-axis ticks to match bin edges

# Show plot
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Age group edges
bins = [20, 30, 40, 50, 60]

# Number of employees in each age group
employees = [13, 24, 6, 4]

# Colors for each bin
colors = ['blue', 'orange', 'green', 'red']

# Plot each age group as a separate histogram
for i in range(len(employees)):
    plt.hist([bins[i]], bins=[bins[i], bins[i+1]], weights=[employees[i]],
             color=colors[i], edgecolor='black', alpha=0.8, rwidth=0.8, label=f'Age Group {bins[i]}-{bins[i+1]}')

# Adding labels and title
plt.xlabel('Age Groups')
plt.ylabel('Number of Employees')
plt.title('Histogram of Employee Age Groups')
plt.xticks(bins)  # Setting x-axis ticks to match bin edges

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
import pandas as pd
from scipy import stats as s
# Adjusted data
score = {
    'Physics': [62, 47, 55, 74, 31, 55, 85, 63, 42, 32, 71, 55],
    'Maths': [89, 8, 7, 67, 55, 47, 72, 76, 79, 44, 92, 99]  # Truncated to match the length
}

df = pd.DataFrame(score)
print(df)
# Arthmetic Mean of the Score Columns in DataFrame
print("\n\n Arithmetic Mean Value in the Distribution")
print("Score1",s.tmean(df["Physics"]).round(2))
print("Score2",s.tmean(df["Maths"]).round(2))

# Harmonic Mean of the Score Columns in DataFrame
print("\n\n Harmonic Mean Value in the Distribution")
print("Score1",s.hmean(df["Physics"]).round(2))
print("Score2",s.hmean(df["Maths"]).round(2))

# Geometric Mean of the Score Columns in DataFrame
print("\n\n Geometric Mean Value in the Distribution")
print("Score1",s.gmean(df["Physics"]).round(2))
print("Score2",s.gmean(df["Maths"]).round(2))
#  Median of the Score Columns in DataFrame
print("\n\n  Median Value in the Distribution")
print("Score1",(df["Physics"]).median())
print("Score2",(df["Maths"]).median())
#  Mode of the Score Columns in DataFrame
print("\n\n  Mode Value in the Distribution")
print("Score1",(df["Physics"]).mode())
print("Score2",(df["Maths"]).mode())



```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Load dataset (Iris dataset)
iris = datasets.load_iris()

# Convert to DataFrame for easier manipulation (optional)
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])

# We will only classify two classes (Binary Classification - Setosa vs Non-Setosa)
df = df[df['target'] != 2]  # Keep only two classes for binary classification
X = df[['sepal length (cm)', 'sepal width (cm)']]  # Select two features for visualization
y = df['target']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the SVM model (using linear kernel here)
svm_model = SVC(kernel='linear')

# Train the model
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:\n", classification_report(y_test, y_pred))

# Visualizing the decision boundary
def plot_svm_decision_boundary(model, X, y):
    # Create a mesh grid based on the data
    h = 0.01  # step size in the mesh
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Plot decision boundary
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)

    # Plot training points
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=50, cmap=plt.cm.coolwarm, edgecolors='k')

    # Highlight support vectors
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                s=100, facecolors='none', edgecolors='k', label='Support Vectors')

    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.title('SVM Decision Boundary with Support Vectors')
    plt.legend()
    plt.show()

# Plot the decision boundary for the linear SVM
plot_svm_decision_boundary(svm_model, X, y)
```

```python
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
from nltk.stem import WordNetLemmatizer, PorterStemmer
# Task 1: Language Identification (Basic Identification using characters)
def language_identification(text):
    # Assume language based on common English characters/words
    if any(char.isalpha() for char in text):
        return "English (Assumed)"
    return "Unknown Language"

# Task 2: Tokenization
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens


# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('stopwords')
nltk.download('wordnet')



# Task 3: Part-of-Speech (POS) Tagging
def pos_tagging(tokens):
    pos_tags = pos_tag(tokens)
    return pos_tags

# Task 4: Chunking (Named Entity Recognition)
def chunking(pos_tags):
    chunks = ne_chunk(pos_tags)
    return chunks

# Task 5: Removing Stop Words
def remove_stopwords(tokens):
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

# Task 6: Syntax Parsing (simulated using POS tags)
def syntax_parsing(pos_tags):
    return [(word, tag) for word, tag in pos_tags]

# Task 7: Lemmatization
def lemmatization(tokens):
    lemmatized_words = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_words

# Task 8: Stemming
def stemming(tokens):
    stemmed_words = [ps.stem(word) for word in tokens]
    return stemmed_words
# Sample text
text = "John, who is an engineer, went to the market to buy fresh vegetables."
# Process the text
tokens = tokenize_text(text)
pos_tags = pos_tagging(tokens)
# Initialize NLTK components
lemmatizer = WordNetLemmatizer()
ps = PorterStemmer()
# Print results for each task
print("Language Identification:", language_identification(text))
print("Tokens:", tokens)
print("Filtered Tokens (Stop Words Removed):", remove_stopwords(tokens))
print("Lemmatized Words:", lemmatization(tokens))
print("Stemmed Words:", stemming(tokens))
#POS_TAG USED
print("POS Tags:", pos_tags)
print("Chunks (NER):", chunking(pos_tags))
print("Syntax Parsing (POS):", syntax_parsing(pos_tags))

```
